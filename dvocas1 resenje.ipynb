```python
# Napraviti program koji prihvata broj i proverava parnost unetog broja
broj = int(input("Unesite broj: "))
if broj == 0:
    print("Broj je nula!")
elif broj % 2 == 0:
    print("Broj je paran!")
else:
    print("Broj je neparan!")
```

```python
# Napraviti program koji omogućava korisnicima unos brojeva i ispisuje sve pozitivne brojeve koje korisnik unese. Negativni brojevi se ne ispisuju, a ako korisnik unese 0, program se prekida.
for _ in range(100):
    broj = int(input("Unesite broj: "))
    if broj < 0:
        continue
    elif broj == 0:
        print("Kraj programa.")
        break
    else:
        print("Pozitivan broj: ", broj)
```

```python
# Napraviti program koji izračunava faktorijel unetog broja
broj = int(input("Unesite broj: "))
faktorijel = 1
for i in range(1,broj+1):
    faktorijel *= i
print(f"Faktorijel broja {broj}: {faktorijel}")
```

```python
# Operacije nad stringovima

# Napisati program koji prikazuje najdužu reč u rečenici
tekst = "Napisati program koji ispisuje najduzu rec u ovoj recenici."
najduza = ""
reci = tekst.split()
for rec in reci:
    if len(rec) > len(najduza):
        najduza = rec
print("Najduza rec: ", najduza)

# Substringovi
print("Vestacka inteligencija"[5:])
print("Vestacka inteligencija"[5:-3])

# Uklanjanje praznih karaktera sa početka i kraja stringa
x = "     Vestacka inteligencija     "
print(x)
x = x.strip()
print(x)

# Zamena karaktera u stringu
print(x.replace("a", "A"))

# Ispis stringa malim ili velikim slovima
print(x.lower())
print(x.upper())
```

```python
# Rečnik (dictionary) je struktura podataka koja čuva parove ključ-vrednost. Ključevi moraju biti jedinstveni.
studenti = {'Ana': 10, 'Milos': 8, 'Jovan':9}
print(studenti)

# Pristup vrednosti preko ključa
print("Anina ocena je: ", studenti['Ana'])

# Dodavanje novog para ključ vrednost
studenti['Milica'] = 10
print(studenti)

# Provera da li ključ postoji u rečniku
if 'Jovan' in studenti:
    print("Jovan je vec u recniku!")
```

```python
# Primer za defaultdict
# defaultdict omogućava kreiranje rečnika gde možemo unapred definisati podrazumevanu vrednost za ključeve koji ne postoje
from collections import defaultdict
# Kreirati defaultdict gde je podrazumevana vrednost lista
ocene_po_predmetima = defaultdict(list)
# Dodavanje ocena po predmetima
ocene_po_predmetima['Matematika'].append(10)
ocene_po_predmetima['Matematika'].append(8)
ocene_po_predmetima['Fizika'].append(9)
print(ocene_po_predmetima)
```

```python
# Primer za tuple (uređeni niz vrednosti)
# Tuple je sličan listi, ali je nepromenljiv (immutable)

# Primer geografskih koordinata
koordinate = (45.76, 19.2)
print("Geografska sirina: ", koordinate[0])
print("Geografska duzina: ", koordinate[1])

# Tuple se može koristiti kao ključ u rečniku jer je nepromenljiv
gradovi = {
    ('Beograd', 'Srbija'): 1400000,
    ('Novi Sad', 'Srbija'): 400000
}
print(gradovi)
```

```python
# Primer za set (skup)
# Set je kolekcija jedinstvenih elemenata. Koristi se kada želimo da eliminišemo duplikate ili da radimo sa jedinstvenim vrednostima. Ponavljanja se automatski uklanjaju.
brojevi = {1, 2, 2, 5, 7, 5, 3}
print(brojevi)

# Dodavanje elemenata u set
brojevi.add(5) #vec postoji
print(brojevi)

# Provera pripadnosti u setu
print("Da li je 3 u setu? ", 3 in brojevi)
```

```python
# Primer za dictionary comprehension
# Dictionary comprehension je način za kreiranje rečnika koristeći sažetu sintaksu.

# Kreirati rečnik koji mapira brojeve od 1 do 5 u njihove kvadrate.
kvadrati = {x: x**2 for x in range(1,6)}
print(kvadrati)
```

```python
# Klase
# Napraviti klasu Temperatura koja konvertuje farenhajte u celzijuse i obrnuto.
class Temperatura:
    def  __init__ (self, celzijus = 0, farenhajt = 0):
        self._celzijus = celzijus
        self._farenhajt = farenhajt
    def get_farenhajt(self):
        return self._celzijus * 9 /5 +32
    def get_celzijus(self):
        return (self._farenhajt - 32) * 5 / 9

t1 = Temperatura (celzijus = 10)
print(isinstance(t1, Temperatura))
print("Temperatura u farenhajtima: ", t1.get_farenhajt())
```

```python
# Instanciranje klasa i definisanje funkcija
class Zaposleni:
    def __init__(self, imePrezime, plata):
        self.imePrezime = imePrezime
        self.plata = plata
    def vratiPrezime(self):
        return self.imePrezime.split(" ")[1]
    def dodajPovisicu(self, procenat):
        self.plata = self.plata * (1.0 + procenat)

z1 = Zaposleni("Pera Peric", 70000)
z1.dodajPovisicu(0.1)
print("Nova plata: ", z1.plata)
```
